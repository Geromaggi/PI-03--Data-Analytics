# Hora

```python
# analicemos la columna 'HORA declarada'
df['hora']
```

```python
# Ahora reemplazeemos los valores nulos por la hora cero. (00:00)
df['hora'].replace(np.nan , '00:00',inplace=True)
```

```python
# Vemos que esta columna, no tiene el formato de hora comun. Para esto crearemos la siguiente funcion:
def normalizar_hora (valor):
    if ':' not in valor:
        return f'{valor[:2]}:{valor[-2:]}'
    else:
        return valor
```

```python
# apliquemosle esta funcion a nuestra columna:
df['hora'] = df['hora'].apply(lambda x: normalizar_hora(x))
```

Encontramos una libreria muy buena llamada "Flydenity". Podemos ver de que se trata y cosas especificas en el siguiente link:

https://github.com/Collen-Roller/flydenity

```python
from flydenity import Parser
parser = Parser()
parser.parse('F-ALTA')['nation'] # aqui vemos un ejemplo de como funciona
```

```python
# creemos la siguiente funcion:
def nation (value):
    try:
        return parser.parse(value)['nation']
    except:
        return 'Desconocido'
```

```python
# ahora agreguemos una columna a nuestro df original y le apliquemos esa funcion:
df['nacionalidad'] = df['codigo_registracion'].dropna().apply(nation)
```

```python
df['nacionalidad'].replace(np.nan, 'Desconocido',inplace=True)
```

A estos dataframe se le han realizado diferentes transformaciones (estas estaran en el archivo datasets_adicionales.ipynb). Luego de estas fueron transformados cada uno a csv y asi es como los traemos aqui para juntarlos con nuestro df original

```python
air_traffic_errors = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\Air_traffic_errors.csv',index_col=0)
```

```python
cargo_hold = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\cargo_hold.csv',index_col=0)
```

```python
design_flaws = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\design_flaws.csv',index_col=0)
```

```python
fuel_starvation = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\fuel_starvation.csv',index_col=0)
```

```python
lightning_accidents = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\lightning_accidents.csv',index_col=0)
```

```python
pilot_incapacitation = pd.read_csv(r'C:\Users\DELL\Desktop\Gero\PI 03\Datasets\pilot_incapacitation.csv',index_col=0)
```

```python
air_traffic_errors['ubicacion'] = air_traffic_errors['ubicacion'].str.replace('Off ','')
cargo_hold['ubicacion'] = cargo_hold['ubicacion'].str.replace('Off ','')
design_flaws['ubicacion'] = design_flaws['ubicacion'].str.replace('Off ','')
fuel_starvation['ubicacion'] = fuel_starvation['ubicacion'].str.replace('Off ','')
lightning_accidents['ubicacion'] = lightning_accidents['ubicacion'].str.replace('Off ','')
pilot_incapacitation['ubicacion'] = pilot_incapacitation['ubicacion'].str.replace('Off ','')
```

```python
# aqui realizaremos una copia de nuestro df origninal
df_copy = df.copy()
```

```python
# creamos una nueva columna que nos diga solo el pais
df_copy['pais'] = df_copy['ruta'].str.split(',').str[-1]
```

```python
df_copy['pais']
```

```python
# para terminar hacemos lo siguiente:
df_copy = df_copy.merge(air_traffic_errors, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
df_copy = df_copy.merge(cargo_hold, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
df_copy = df_copy.merge(design_flaws, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
df_copy = df_copy.merge(fuel_starvation, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
df_copy = df_copy.merge(lightning_accidents, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
df_copy = df_copy.merge(pilot_incapacitation, how='left', left_on=['fecha', 'pais'], right_on = ['fecha', 'ubicacion']).drop('ubicacion', axis=1)
```

```python
# rellenamos con 'No' los nulos.
df_copy['air_traffic_errors'].fillna('No', inplace=True)
df_copy['cargo_hold'].fillna('No', inplace=True)
df_copy['design_flaws'].fillna('No', inplace=True)
df_copy['fuel_starvation'].fillna('No', inplace=True)
df_copy['lightning_accidents'].fillna('No', inplace=True)
df_copy['pilot_incapacitation'].fillna('No', inplace=True)
```

```python
# agregamos una ultima columna:
df_copy['total_fallecidos(+ tierra)'] = df_copy['total_fallecidos'] + df_copy['tierra']
```
