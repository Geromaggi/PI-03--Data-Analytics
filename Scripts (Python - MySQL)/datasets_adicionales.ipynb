```python
import pandas as pd
```

Aqui traeremos todos estos dataframes adicionales, se la haran las mismas transformaciones a todos. 
Leemos, dropeamos nulos, la columna fecha la pasamos a tipo date y le ponemos el formato correcto

```python
air_traffic_errrors = pd.read_html('http://www.planecrashinfo.com/CATATC.htm')
air_traffic_errrors = air_traffic_errrors[0].dropna().drop(0, axis=0).drop(3,axis=1)
air_traffic_errrors[0] = pd.to_datetime(air_traffic_errrors[0],format='%m/%d/%Y')
air_traffic_errrors
```

```python
cargo_hold = pd.read_html('http://www.planecrashinfo.com/CATCH.htm')
cargo_hold = cargo_hold[0].dropna().drop(0, axis=0).drop(3,axis=1)
cargo_hold[0] = pd.to_datetime(cargo_hold[0],format='%m/%d/%Y')
cargo_hold
```

```python
design_flaws = pd.read_html('http://www.planecrashinfo.com/CATDF.htm')
design_flaws = design_flaws[0].dropna().drop(0, axis=0).drop(3,axis=1)
design_flaws[0] = pd.to_datetime(design_flaws[0],format='%m/%d/%Y')
design_flaws
```

```python
fuel_starvation = pd.read_html('http://www.planecrashinfo.com/CATFS.htm')
fuel_starvation = fuel_starvation[0].dropna().drop(0, axis=0).drop(3,axis=1)
fuel_starvation[0] = pd.to_datetime(fuel_starvation[0],format='%m/%d/%Y')
fuel_starvation
```

```python
lightning_accidents = pd.read_html('http://www.planecrashinfo.com/CATL.htm')
lightning_accidents = lightning_accidents[0].dropna().drop(0, axis=0).drop(3,axis=1)
lightning_accidents[0] = pd.to_datetime(lightning_accidents[0],format='%m/%d/%Y')
lightning_accidents
```

```python
pilot_incapacitation = pd.read_html('http://www.planecrashinfo.com/CATPI.htm')
pilot_incapacitation = pilot_incapacitation[0].dropna().drop(0, axis=0).drop(3,axis=1)
pilot_incapacitation[0] = pd.to_datetime(pilot_incapacitation[0],format='%m/%d/%Y')
pilot_incapacitation
```

```python
dict_names = {0:'fecha', 1:'ubicacion', 2:'operador'} #nombre que le pondremos a cada columna
```

```python
# ahora les cambiaremos el nombre a todos nuestros df
air_traffic_errrors.rename(columns=dict_names, inplace=True)
cargo_hold.rename(columns=dict_names, inplace=True)
design_flaws.rename(columns=dict_names, inplace=True)
fuel_starvation.rename(columns=dict_names, inplace=True)
lightning_accidents.rename(columns=dict_names, inplace=True)
pilot_incapacitation.rename(columns=dict_names, inplace=True)
```

```python
#dropeamos nulos
air_traffic_errrors.dropna(inplace=True)
cargo_hold.dropna(inplace=True)
design_flaws.dropna(inplace=True)
fuel_starvation.dropna(inplace=True)
lightning_accidents.dropna(inplace=True)
pilot_incapacitation.dropna(inplace=True)
```

```python
#agregamos una columna
air_traffic_errrors['air_traffic_errors'] = 'Si'
cargo_hold['cargo_hold'] = 'Si'
design_flaws['design_flaws'] = 'Si'
fuel_starvation['fuel_starvation'] = 'Si'
lightning_accidents['lightning_accidents'] = 'Si'
pilot_incapacitation['pilot_incapacitation'] = 'Si'
```

```python
#modificamos bien los espacios despues de la coma
air_traffic_errrors['ubicacion'] = air_traffic_errrors['ubicacion'].str.replace(', ', ',')
cargo_hold['ubicacion'] = cargo_hold['ubicacion'].str.replace(', ', ',')
design_flaws['ubicacion'] = design_flaws['ubicacion'].str.replace(', ', ',')
fuel_starvation['ubicacion'] = fuel_starvation['ubicacion'].str.replace(', ', ',')
lightning_accidents['ubicacion'] = lightning_accidents['ubicacion'].str.replace(', ', ',')
pilot_incapacitation['ubicacion'] = pilot_incapacitation['ubicacion'].str.replace(', ', ',')
```

```python
# dejamos solamente el pais
air_traffic_errrors['ubicacion'] = air_traffic_errrors['ubicacion'].str.split(',').str[-1]
cargo_hold['ubicacion'] = cargo_hold['ubicacion'].str.split(',').str[-1]
design_flaws['ubicacion'] = design_flaws['ubicacion'].str.split(',').str[-1]
fuel_starvation['ubicacion'] = fuel_starvation['ubicacion'].str.split(',').str[-1]
lightning_accidents['ubicacion'] = lightning_accidents['ubicacion'].str.split(',').str[-1]
pilot_incapacitation['ubicacion'] = pilot_incapacitation['ubicacion'].str.split(',').str[-1]
```

```python
# dropeamos la columna operador
air_traffic_errrors.drop('operador', axis=1, inplace=True)
cargo_hold.drop('operador', axis=1, inplace=True)
design_flaws.drop('operador', axis=1, inplace=True)
fuel_starvation.drop('operador', axis=1, inplace=True)
lightning_accidents.drop('operador', axis=1, inplace=True)
pilot_incapacitation.drop('operador', axis=1, inplace=True)
```

```python
# ahora pasamos cada uno a csv:
air_traffic_errrors.to_csv('data/Air_traffic_errors.csv')
cargo_hold.to_csv('data/cargo_hold.csv')
design_flaws.to_csv('data/design_flaws.csv')
fuel_starvation.to_csv('data/fuel_starvation.csv')
lightning_accidents.to_csv('data/lightning_accidents.csv')
pilot_incapacitation.to_csv('data/pilot_incapacitation.csv')
```
